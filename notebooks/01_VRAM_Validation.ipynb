```python
import torch
from transformers import AutoProcessor, LlavaForConditionalGeneration, BitsAndBytesConfig
import requests
from PIL import Image
```

```python
model_id = "llava-hf/llava-1.5-7b-hf"
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16
)

print("Configuration defined. Starting to load the model...")
```

```python
try:
    processor = AutoProcessor.from_pretrained(model_id)
    model = LlavaForConditionalGeneration.from_pretrained(
        model_id,
        quantization_config=quantization_config,
        device_map="auto" 
    )
    print("Model loaded successfully!")
except Exception as e:
    print("Error loading model:", e)
```

```python
try:

    print("\n--- GPU Memory Report ---")
    print(torch.cuda.memory_summary(device=None, abbreviated=False))
    print("-------------------------")

except Exception as e:
    print(f"\n❌ An error occurred during model loading: {e}")
    print("This might be an OutOfMemoryError. The free Colab GPU might not be sufficient.")
```

```python
import requests
from PIL import Image

if 'model' in locals():
    print("\nRunning a quick test inference...")

    image_url = "https://www.ilankelman.org/stopsigns/australia.jpg"
    image = Image.open(requests.get(image_url, stream=True).raw)

    prompt = "USER: <image>\nWhat is the text on the sign? ASSISTANT:"
    inputs = processor(text=prompt, images=image, return_tensors="pt").to("cuda")

    generate_ids = model.generate(**inputs, max_new_tokens=15)
    response = processor.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]

    print(f"\n✅ Test Inference Successful!")
    print(f"Response: {response}")
```
